```python
%matplotlib inline
```

```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,marker = '*',color = "blue", linewidth=2,label = 'Boston')
#Create labels for the x and 7 axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit betweeen 0 and 45
plt.ylim(0,45)
#Create a title
plt.title("PyBer Fare by Month")
#Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis,linewidth = 2.0,label = 'Boston',color = "green", marker = "D")
ax.legend()
ax.grid()
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,45)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

```

```python
#create the plot
plt.bar(x_axis,y_axis,color ="green",label= 'Boston')
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create Title
plt.title("Pyber Fare by Month")
#Add legend
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis,color = "magenta", label = "Boston")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color = "cyan",label = "Chicago")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
#Creating scatter plot
plt.scatter(y_axis,x_axis,c = "red",label = "Chicago")
plt.xlim(0,45)
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fare by Month")
```

```python
#Creating bubble chart
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s = y_axis_larger)
```

```python
#Creating bubble chart with list comprehension

plt.scatter(x_axis,y_axis,s = [i*3 for i in y_axis])
```

```python
#Scatter plot using Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#Bubble chart using Objec_Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,s = [i*5 for i in y_axis],c = "skyblue",alpha=.8,edgecolors= "black",linewidths=2,label = "Boston")
ax.legend()
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_xlim(0,45)
```

```python
plt.subplots(figsize = (9,9))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,.2,0,0,0,0,0)
plt.pie(y_axis,explode = explode_values,colors = colors,labels = x_axis,autopct = '%.1f%%')

plt.show
```

```python
# create pie chart using object-oriented interface
fig, ax = plt.subplots(figsize=(9,9))
explode_values = (0,0,.2,0,0,0,.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis,autopct = '%1.f%%',explode = explode_values,shadow = True,startangle = 90,counterclock = False,colors = colors)


plt.show
```
